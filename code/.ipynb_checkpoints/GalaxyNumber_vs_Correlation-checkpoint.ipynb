```python
import numpy as np
import matplotlib.pyplot as plt
import LSSfunctions as lssfunc
import GSMfunctions as gsmfunc
import StatsFunctions as statsfunc
```

## Is galaxy number affecting correlation?

Theoretically, it shouldn't. Indeed, suppose $x, y$ are two random variables, and $a$ is a scalar. Clearly,<br>
$E(a \cdot x) = a \cdot E(x)$, <br>
and thus, <br>
$STD(a \cdot x) = \sqrt{E(a^{2} \cdot x^{2}) - E(a \cdot x)^{2}} = \sqrt{a^{2} \cdot (E(x^{2}) - E(x)^{2})} = a \cdot STD(x)$. <br>
Consequently, 

$COR(a \cdot x,y) = \frac{COV(a \cdot x, y)}{STD(a \cdot x) \cdot STD(y)} = \frac{E(a \cdot xy) - E(a \cdot x)E(y)}{a \cdot STD(x)STD(y)} = \frac{a \cdot E(xy) - a \cdot E(x)E(y)}{a \cdot STD(x)STD(y)} = COR(x,y)$. <br>

Applying this to our case, if x represents the LSS data and y the temperature data, then scaling the LSS data, i.e. increasing the number of galaxies by keeping the same structural aspects of the data, should not affect the correlation obtained.


We check this by plotting a correlation vs z graph, where we will scale the LSS data by a factor 'a' :

```python
def scaling_uniform_plot(sample_type, div=200, log=True, numpoints=15, centered=False):
    z = np.linspace(0,0.5,numpoints)
    temp = gsmfunc.cctemp(log=log, freq=500, div=div, centered=centered)
    x = (z[1:] + z[:-1])/2
    
    plt.figure(figsize=(12,8))
    plt.xlabel('z')
    plt.ylabel('Corr')
    for a in [0.1, 1, 2, 10, 100]:
        y = []
        for i in range(len(z)-1):
            lssdata = lssfunc.cclss(sample_type, zmin=z[i], zmax=z[i+1], div=div, centered=centered) #We get the data
            lssdata_scaled = np.multiply(lssdata,a)
            corr,_ = statsfunc.correlate(temp, lssdata_scaled, subarr_size=50) #We compute the correlation
            y.append(corr)
        plt.plot(x,y,linestyle='--', label = 'a = '+str(a))

    plt.legend()
```

They overlap, and thus are identical, as expected.

```python
scaling_uniform_plot('bright')
```

### Conclusion: 
The number of galaxies in each z-range is not affecting the correlation function output. Only the patterns and structure in the data is.

### Further questioning on the topic:

Suppose the number of galaxies is not scaling uniformly, that is, the number of galaxies DOES affect the correlation, but not by uniform scaling. Instead, it affects it by scaling density in regions by different values that are within a close range of some scaling factor $\beta$. In other words, we are adding a noise to our scaling to observe how it affects the correlation. 

So the questions we have are:

How is non-uniform scaling going to affect the data? Is the structure of the data solid enough to whithold small discrepancies in scaling? Is the data structured enough for us to conclude that non-zero correlation is justified?

```python
def scaling_non_uniform_plot(sample_type, div=200, log=True, numpoints=15, centered=False):
    z = np.linspace(0,0.5,numpoints)
    temp = gsmfunc.cctemp(log=log, freq=500, div=div, centered=centered)
    x = (z[1:] + z[:-1])/2
    
    plt.figure(figsize=(12,8))
    plt.xlabel('z')
    plt.ylabel('Corr')
    for margin in [0, 0.01, 0.05, 0.1, 0.2, 0.5, 1]:
        #scaling_matrix = np.random.uniform(1-margin, 1+margin, size = (div,div)) #We initialize the scaling matrix
        scaling_matrix = np.random.normal(loc = 1, scale = margin, size=(div,div))
        y = []
        error = []
        for i in range(len(z)-1):
            lssdata = lssfunc.cclss(sample_type, zmin=z[i], zmax=z[i+1], div=div, centered=centered) #We get the data
            lssdata_scaled = np.multiply(lssdata, scaling_matrix) #We scale non-uniformly the LSS data
            corr,err = statsfunc.correlate(temp, lssdata_scaled, subarr_size=50) #We compute the correlation
            y.append(corr)
            error.append(err)
            
        if margin == 0:
            plt.plot(x,y, color='red', lw=2, label = 'no scaling')
            plt.errorbar(x,y, yerr=error,fmt='.k', color='black', capsize=5, label = 'error when no scaling')
        else:
            plt.plot(x,y,linestyle='--', label = 'scaling factor in range '+str(1-margin) + ' to ' + str(1+margin))

    plt.legend()
```

When performing non-uniform scaling on the data, we see that the correlation curve maintains its structure. Intuitively, this could occur when an image lacks structural integrity. Indeed, if randomly scaling pixel values all around the image does not change a lot how the image correlates with another image, it can illustrate that the image itself does not have well defined structures.

```python
scaling_non_uniform_plot('bright')
```

Lets explore this mathematically.

Suppose x,y,z are random variables, where z is drawn from a gaussian distribution. Moreover, suppose z is independent of x and y. Then, <br><br>
$COV(x \cdot z, y) = E(xyz) - E(xz)E(y) = E(z) \cdot COV(x,y)$, and <br>
$VAR(xz) = E(x^{2}z^{2}) - E(xz)^{2} = E(x^{2})E(z^{2}) - E(x)^{2}E(z)^{2} = E(x^{2}) \cdot (E(z)^{2} + VAR(z)) - E(x)^{2}E(z)^{2} = E(z)^{2} \cdot VAR(x) + E(x^{2}) \cdot VAR(z)$ <br><br>

Thus, 

$COR(x \cdot z, y) = \frac{E(z) \cdot COV(x,y)}{\sqrt{VAR(y) \cdot \big(E(z)^{2} \cdot VAR(x) + E(x^{2}) \cdot VAR(z) \big)}}$<br>

In our specific case, we have that $E(z) = 1$ and $VAR(z)$ is the margin squared. Thus, if $m$ is the margin,

$COR(x \cdot z, y) = \frac{COV(x,y)}{\sqrt{VAR(y) \cdot \big(VAR(x) + m^{2}E(x^{2}) \big)}} = COR(x,y) \cdot \sqrt{\frac{VAR(x)}{VAR(x) + m^{2}E(x^{2})}} = COR(x,y) \cdot \sqrt{\frac{1}{1 + \frac{m^{2}E(x^{2})}{VAR(x)}}} = COR(x,y) \cdot \sqrt{\frac{1}{1 + m^{2} \alpha}}$,<br>

where $\alpha = \frac{E(x^{2})}{VAR(x)} = 1 + \frac{E(x)^{2}}{VAR(x)} = 1 + \big(\frac{E(x)}{STD(x)}\big)^{2}$. <br>

Note: We assume VAR(x) is not 0.


Coming back to our plot, this makes perfect sense because increasing margins lead to smaller correlation values. Indeed, in our plot we fix $\alpha$ and change the margin for each line, and the progression is clear.

It is also important to mention why the choice of a distribution centered on 1. This is because centering it on any other value such as 7 is equivalent to uniformly scaling the data by 7, and then non-uniformly scaling it. We know from previous observations that uniform scaling does not affect correlation. So it is clear that the former step is unecessary.

```python
def alpha(x):
    return 1 + (x.mean()/x.std())**2
```
