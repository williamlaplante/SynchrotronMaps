```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

import StatsFunctions as statsfunc
```

## Exploring effects of subarray size on correlation for Gaussian samples
Let $A,B$ be two arrays generated by gaussian sampling.

More precisely, we observe how averaging over correlation on subarrays of $A,B$ of different size compares with the correlation applied on $A,B$.

```python
def corr_gauss(meanA=0, stdA=0.1, meanB=1, stdB=0.2):
    
    #Fixing the size and subarray sizes
    size = 5*4*3*2*1
    subarr_sizes = [2,3,4,5,6,8,10,12,15,20,30,60,120] 
    x = [size**2 for size in subarr_sizes]
    
    #Generating the matrices

    A = np.random.normal(loc = meanA, scale = stdA, size=(size,size))
    B = np.random.normal(loc = meanB, scale = stdB, size=(size,size))
    
    #Obtaining the average correlation values for multiple subarray sizes and the associated error
    corr_values = []
    corr_error = []
    for subarr_size in subarr_sizes:
        corr, err = statsfunc.correlate(A,B,subarr_size=subarr_size, filter_zeros=False)
        corr_values.append(corr)
        corr_error.append(err)
    
    #Plotting the correlation and the error
    fig, ax = plt.subplots(1, 2, figsize=(14,6))
    ax[0].plot(x, corr_values, color='blue',marker='o', label='<COR(X_i, Y_i)>')
    ax[0].set_xlabel('subarray size')
    ax[0].set_ylabel('Average correlation')
    ax[1].plot(x, corr_error, color='orange')
    ax[1].set_xlabel('subarray size')
    ax[1].set_ylabel('Error of average correlation')
    ax[0].plot(x,[statsfunc.corrfunc(A,B)]*len(x), '--', color='orange', label='COR(X,Y)')
    
    #Fitting the error
    def func(n,a): return a/n
    popt, pcov = curve_fit(func, x, corr_error)
    ax[1].plot(x, func(x, *popt), label='a/n fit with a = '+str(popt[0]))
    
    def func2(n,a): return a/np.sqrt(n)
    popt, pcov = curve_fit(func2, x, corr_error)
    ax[1].plot(x, func2(x, *popt),color='red', label='a/sqrt(n) fit with a = '+str(popt[0]))
    ax[1].set_xscale('log')
    ax[0].set_xscale('log')
    ax[1].legend()
    ax[0].legend()
    return 
```

```python
corr_gauss()
```

### Explanation

We define the average correlation when the subarray size is of $n$ as follows:

$ <COR(X_{i},Y_{i})>_{n} = <\frac{COV(X_{i},Y_{i})}{\sigma(X_{i}) \sigma(Y_{i})}>_{n} = <\frac{E(X_{i} Y_{i}) - E(X_{i})E(Y_{i})}{\sigma(X_{i}) \sigma(Y_{i})}>_{n}$. 

Now, we expect the correlation of the submatrices to be close to the correlation of the whole arrays, up to some constant that should depend on $n$.

$ <COR(X_{i},Y_{i})>_{n} \approx \delta _{n} + COR(X,Y)$

On the other hand, following a similar reasoning,

$ <COR(X_{i}, Y_{i})> \approx \frac{n}{m} \sum_{i=1}^{m/n} COR(X_{i}, Y_{i}) \approx COR(X,Y) + \frac{n}{m}\sum_{i=1}^{m/n} \Delta _{n,i}$

But the $\Delta _{n,i}$ each represent the error(std) in the correlation of the subarrays for some fixed subarray size $n$, which, when considering the subarrays as samples of the whole arrays (population), should depend on their size as $\frac{1}{\sqrt{n}}$. 

To see this, we look at the variance of the covariance:

$VAR(COV(X_{i},Y_{i})) = VAR(E(X_{i}Y_{i})) + VAR(E(X_{i})E(Y_{i})) \propto \frac{1}{\sqrt{n}}$, since $X_{i},Y_{i}$ are made of $n$ elements.


Consequently, 

$\delta _{n}  = \frac{n}{m}\sum_{i=1}^{m/n} \Delta _{n,i} \propto \sqrt{\frac{1}{n}}$

Hence, we conclude that 

$ STD(<COR(X_{i},Y_{i})>_{n}) \propto  \sqrt{\frac{1}{n}}$


### Exploring effects of subarray size on correlation for data samples

```python
import GSMfunctions as gsmfunc
import LSSfunctions as lssfunc
```

```python
def corr_data():
    size = 5*4*3*2*1*2
    subarr_sizes = [2,3,4,5,6,8,10,12,15,20,30,40,60,120,240]
    x = [size**2 for size in subarr_sizes]
    
    fig, ax = plt.subplots(1, 2, figsize=(14,6))

    
    A = lssfunc.cclss('bright', zmin=0.14, zmax=0.17, div=size)
    B = gsmfunc.cctemp(div=size)

    corr_values = []
    corr_error = []
    for subarr_size in subarr_sizes:
        corr, err = statsfunc.correlate(A,B,subarr_size=subarr_size, filter_zeros=True)
        corr_values.append(corr)
        corr_error.append(err)

    #Plotting the correlation and the error
    ax[0].plot(x, corr_values, color='blue')
    ax[0].set_xlabel('subarray size')
    ax[0].set_ylabel('Average correlation')
    ax[1].plot(x, corr_error, color='orange')
    ax[1].set_xlabel('subarray size')
    ax[1].set_ylabel('Error of average correlation')

    
    #Fitting the error
    
    def func(n,a): return a/n
    popt, pcov = curve_fit(func, x, corr_error)
    ax[1].plot(x, func(x, *popt), label='a/n fit with a = '+str(popt[0]))
    
    def func2(n,a): return a/np.sqrt(n)
    popt, pcov = curve_fit(func2, x, corr_error)
    ax[1].plot(x, func2(x, *popt),color='red', label='a/sqrt(n) fit with a = '+str(popt[0]))
    
    ax[1].legend()
    
    return
```

```python
corr_data()
```

```python
def corr_data_z():
    
    size = 5*4*3*2*1
    subarr_sizes = [2,3,4,5,6,8,10,12,15,20,30,40,60,120]
    x = [size**2 for size in subarr_sizes]
    z = np.linspace(0,0.5,10)
    plt.figure(figsize=(12,10))
    
    for i in range(len(z)-1):
        
        A = lssfunc.cclss('bright', zmin=z[i], zmax=z[i+1], div=size)
        B = gsmfunc.cctemp(div=size)

        corr_values = []
        corr_error = []
        
        for subarr_size in subarr_sizes:
            corr, err = statsfunc.correlate(A,B,subarr_size=subarr_size, filter_zeros=True)
            corr_values.append(corr)
            corr_error.append(err)
        
        plt.plot(x, corr_values, linestyle='--', marker='o', label= str(round(z[i],3)) + ' < z < ' + str(round(z[i+1],3)))
    
    plt.xlabel('subarray size')
    plt.ylabel('correlation')
    plt.legend()
    return 
```

From the figure, we see that for most z-ranges, there is a point of lower subarray size which reflects almost the same value as the correlation value for the full array. Consequently, a way to proceed to quote the error would go as follows:

For each z range: <br>
1) Find at which subarray size the correlation value matches the best the correlation value of the whole array.<br>
2) Quote this value for the correlation value in the appropriate z range.<br>
3) Find the discrepancy of the correlation values at the corresponding subarray size and quote it as the error.<br>

```python
corr_data_z()
```
