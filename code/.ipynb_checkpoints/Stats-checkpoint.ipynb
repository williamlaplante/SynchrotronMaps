```python
import GSMfunctions as gsmfunc
import LSSfunctions as lssfunc
import Statsfunctions as statsfunc
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def zinfo():
    '''
    Just to get boundaries on z values for each catalog
    
    '''
    sample_types = ['bright', 'void', 'safe', 'full']
    df = pd.DataFrame([], index=['zmin','zmax'])
    for sample_type in sample_types : 
        data = lssfunc.getLSSdata(sample_type=sample_type)
        zmin = data['Z'].min()
        zmax = data['Z'].max()
        df[sample_type] = [zmin, zmax]
    return df
zinfo()
```

```python
def visualizetemp():
    plt.figure(figsize=(8,8))
    plt.imshow(gsmfunc.cctemp(200, freq=500))
    plt.colorbar()
    plt.axis('off')
    return 
visualizetemp()
```

```python
def visualizelss(sample_type):
    plt.figure(figsize=(8,8))
    plt.imshow(lssfunc.cclss(sample_type, zmin=0.1, zmax=0.2, div=200))
    plt.colorbar()
    plt.axis('off')
    return
visualizelss('safe')
```
